## Import data

```python
# import the scrubbed and tagged data
df = pd.read_csv('responses_scrubbed_tagged.csv', index_col=None, parse_dates=['date'])
```

## Breakdown demographics by neighborhood

```python
# get just the neighborhoods and demographic
df_hoods = df[['neighborhood', 'demographic']]

# split different values into diff
dummies = pd.get_dummies(df_hoods['demographic'])
#dummies = dummies.rename(columns=lambda x: 'Category_' + str(x))
df_hoods = pd.concat([df_hoods, dummies], axis=1)
df_hoods.drop(['demographic'], inplace=True, axis=1)

# group
df_hoods = df_hoods.groupby('neighborhood')

# do aggregate
df_hoods = df_hoods.agg('sum')

df_hoods
```

### Number of respondents from different areas of the village

```python
df_hoods.sort_index(ascending=False).plot(
        kind='barh',
        figsize=(10,15),
#        cmap='rocket',
#        edgecolor='black',
        fontsize=14,
        title='Number of respondents from different areas'
    ).yaxis.label.set_visible(False)
```
